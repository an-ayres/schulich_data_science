```python
# Convert 'Sex' to numeric (0 for male, 1 for female)
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
```

```python
# Create a binary feature 'Has_Cabin' where 1 means the passenger had a cabin, and 0 means they didn't
df['Has_Cabin'] = df['Cabin'].apply(lambda x: 0 if pd.isnull(x) else 1)

# Dropping 'Cabin' for 'Has_Cabin' 
df = df.drop('Cabin', axis=1)
```

```python
# Impute missing values with the mode
df['Age'].fillna(df['Age'].mode()[0], inplace=True)
```
