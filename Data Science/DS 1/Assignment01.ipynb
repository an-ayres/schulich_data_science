```python
from scipy import stats

# Conduct t-test for Age between churned and non-churned users
churned_age = df[df['Churned'] == 1]['Age']
non_churned_age = df[df['Churned'] == 0]['Age']

t_statistic, p_value = stats.ttest_ind(churned_age, non_churned_age)

print("T-Statistic:", t_statistic)
print("P-Value:", p_value)
```

```python
# we begin the EDA to know why users are churning

# Plot distributions of key variables
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Age distribution
sns.histplot(data=df, x='Age', hue='Churned', kde=True, ax=axes[0, 0])
axes[0, 0].set_title('Age Distribution by Churn Status')

# Monthly Payment distribution
sns.histplot(data=df, x='Monthly_Payment', hue='Churned', kde=True, ax=axes[0, 1])
axes[0, 1].set_title('Monthly Payment Distribution by Churn Status')

# Monthly Hours Watched distribution
sns.histplot(data=df, x='Monthly_Hours_Watched', hue='Churned', kde=True, ax=axes[1, 0])
axes[1, 0].set_title('Monthly Hours Watched Distribution by Churn Status')

# Average Session Length distribution
sns.histplot(data=df, x='Average_Session_Length', hue='Churned', kde=True, ax=axes[1, 1])
axes[1, 1].set_title('Average Session Length Distribution by Churn Status')

plt.tight_layout()
plt.show()
```

```python
# make it into a table so it's easier to understand

# Calculate summary statistics for churned vs. non-churned users
summary_table = df.groupby('Churned').agg({
    'User_ID': 'count',
    'Age': 'mean',
    'Monthly_Payment': 'mean',
    'Monthly_Hours_Watched': 'mean',
    'Average_Session_Length': 'mean'
}).rename(columns={'User_ID': 'Count', 'Age': 'Average Age', 'Monthly_Payment': 'Average Monthly Payment', 'Monthly_Hours_Watched': 'Average Monthly Hours Watched', 'Average_Session_Length': 'Average Session Length'})

print(summary_table)
```

```python
# exploratory variables regarding patterns and behaviours of the clients that are cancelling their subscription

# Engagement Level
df['Engagement_Level'] = df['Monthly_Hours_Watched'] * df['Average_Session_Length']

print(df['Engagement_Level'])
```

```python
# payment behaviour 
df['Payment_behav'] = df['Monthly_Payment'] / df['Monthly_Hours_Watched']

# Payment Sensitivity
df['Payment_Sensitivity'] = df['Monthly_Payment'] / df['Monthly_Hours_Watched']

# Display the first few rows to verify the new variables
print(df[['Payment_behav', 'Payment_Sensitivity']].head())
```

```python
df['Subscription_Start_Date'] = pd.to_datetime(df['Subscription_Start_Date'])
df['Date_of_Churn'] = pd.to_datetime(df['Date_of_Churn'])

# Average Subscriptions Per Month (Tenure in Months)
df['Tenure'] = (df['Date_of_Churn'] - df['Subscription_Start_Date']).dt.days / 30
average_tenure = df['Tenure'].mean()
print(f'Average Subscriptions Per Month: {average_tenure:.2f} months')

# Average Genre Watched by People Who Churned
genre_churned = df[df['Churned'] == 1]['Favorite_Genre'].value_counts().idxmax()
print(f'Most Watched Genre by People Who Churned: {genre_churned}')

# Average Time Spent on the Platform Before Churning
average_time_spent_before_churning = df[df['Churned'] == 1]['Tenure'].mean()
print(f'Average Time Spent on Platform Before Churning: {average_time_spent_before_churning:.2f} months')

# Devices Used by Users Who Churned
devices_used_by_churned = df[df['Churned'] == 1]['Devices_Used'].value_counts()
print('Devices Used by Users Who Churned:')
print(devices_used_by_churned)

# Age of Users Who Are Churning the Most
most_common_age_churned = df[df['Churned'] == 1]['Age'].mode()[0]
print(f'Age of Users Churning the Most: {most_common_age_churned}')

# Time into the Service Before Users Churn
df['Time_Before_Churn'] = (df['Date_of_Churn'] - df['Subscription_Start_Date']).dt.days
average_time_before_churn_days = df['Time_Before_Churn'].mean()
print(f'Average Time Before Churn: {average_time_before_churn_days:.2f} days')

# Average Time a User Spends with the Service Before Churning
average_time_with_service_before_churning = df[df['Churned'] == 1]['Time_Before_Churn'].mean()
print(f'Average Time a User Spends with the Service Before Churning: {average_time_with_service_before_churning:.2f} days')
```

```python
# Attempting to calculate how users churning is financially affecting the company

# Convert date columns to datetime
df['Subscription_Start_Date'] = pd.to_datetime(df['Subscription_Start_Date'])
df['Date_of_Churn'] = pd.to_datetime(df['Date_of_Churn'])

# Calculate Tenure in months
df['Tenure'] = (df['Date_of_Churn'] - df['Subscription_Start_Date']).dt.days / 30

# Calculate Current Churn Rate
churn_rate = df['Churned'].mean()
print(f'Current Churn Rate: {churn_rate:.2%}')

# Calculate ARPU
arpu = df['Monthly_Payment'].mean()
print(f'Average Revenue Per User (ARPU): ${arpu:.2f}')

# Total Number of Users
total_users = df['User_ID'].nunique()
print(f'Total Number of Users: {total_users}')

# Calculate CLTV
average_tenure_months = df['Tenure'].mean()
cltv = arpu * average_tenure_months
print(f'Customer Lifetime Value (CLTV): ${cltv:.2f}')

# Estimate Financial Impact
estimated_churn_reduction = 0.10  # 10% reduction
retained_users = total_users * churn_rate * estimated_churn_reduction
additional_revenue = retained_users * cltv
print(f'Estimated Additional Revenue from Churn Reduction: ${additional_revenue:.2f}')
```
